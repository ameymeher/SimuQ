```python
import qiskit.circuit.library as circuits
from qiskit import QuantumCircuit
from numpy import pi
import numpy as np
```

```python
#Init parameters of the IBMProvider
api_file = "../../ibm_API_key"
aais = 'heisenberg'
backend = 'ibmq_mumbai'
tol = 0.01
trotter_num = 6
verbose = 0
use_pulse = True
state_prep = None
```

```python
#Init functionality of the IBMProvider
from qiskit_ibm_provider import IBMProvider
with open(api_file, "r") as f:
        api_key = f.readline().strip()

provider = IBMProvider(api_key, instance='ibm-q-ncsu/nc-state/quantum-compiler')
```

```python
backend = provider.get_backend(backend)
```

```python
from qiskit_transpiler import get_pm

pm = get_pm(backend)
```

## RX gate calibration

```python
qc = QuantumCircuit(1)
qc.rx(pi/2,0)
qc.draw('mpl')
```

```python
transpiled_qc = pm.run(qc)
```

```python
transpiled_qc.draw('mpl')
```

```python
transpiled_qc.calibrations
```

```python
from qiskit_experiments.library.randomized_benchmarking import InterleavedRB

lengths = np.arange(1, 200, 30)
num_samples = 10
seed = 1010
qubits = (0,1)

# The interleaved gate is the CX gate
int_exp2 = InterleavedRB(
    transpiled_qc.to_gate(),[0], lengths, num_samples=num_samples, seed=seed)
```

```python
int_exp2.circuits()[3].decompose().draw('mpl',idle_wires=False)
```

```python
int_exp2.circuits()[3].decompose().calibrations
```

```python
int_expdata2 = int_exp2.run(backend).block_for_results()
int_results2 = int_expdata2.analysis_results()
```

```python
# View result data
display(int_expdata2.figure(0))
for result in int_results2:
    print(result)
```
