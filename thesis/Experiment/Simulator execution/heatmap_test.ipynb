```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#Read xlsx file
df = pd.read_excel('results/RBZNE - Linear fit with extra point/summary_distribution_data.xlsx',sheet_name='Test')

#Sort by N in descending order, and then by T in ascending order
df = df.sort_values(by=['N', 'T'], ascending=[True, False])

# Pivot the data to create a 2D heatmap
heatmap_data = df.pivot_table(index='N', columns='T', values='Without Noise HF')

# Plot heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(heatmap_data.sort_values(by='N', ascending=False), cmap='viridis', linewidths=0.2, annot=True, fmt=".4f", cbar_kws={'label': 'Hellinger Fidelity'}, annot_kws={"size": 8})
plt.title('Heatmap of Hellinger Fidelity for Hamiltonian simulation results on noise free simulator', fontsize=10)
plt.xlabel('Time duration of the simulation (T)')
plt.ylabel('Number of qubits (N)')
plt.show()
```

```python
# Pivot the data to create a 2D heatmap
heatmap_data = df.pivot_table(index='N', columns='T', values='Without Noise TV')

# Plot heatmap with inverted color scale
plt.figure(figsize=(10, 8))
sns.heatmap(heatmap_data.sort_values(by='N', ascending=False), cmap='viridis', linewidths=0.2, annot=True, fmt=".4f", cbar_kws={'label': 'Total Variational Distance'}, annot_kws={"size": 8}, vmin=1.0, vmax=0.0)

plt.title('Heatmap of Total Variational Distance for Hamiltonian simulation results on noise free simulator', fontsize=9)
plt.xlabel('Time duration of the simulation (T)')
plt.ylabel('Number of qubits (N)')
plt.show()
```
