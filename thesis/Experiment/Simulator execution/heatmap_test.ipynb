```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#Read xlsx file
df = pd.read_excel('results/RBZNE - Linear fit with extra point/summary_distribution_data.xlsx',sheet_name='Test')
df = df[['N','T','Without Noise HF']]


# Pivot the data to create a 2D heatmap
heatmap_data = df.pivot_table(index='N', columns='T', values='Without Noise HF')

# Plot heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(heatmap_data, cmap='viridis', linewidths=0.5, annot=True, fmt=".4f", cbar_kws={'label': 'Value'})
plt.title('2D Heatmap with Intensity of Color')
plt.xlabel('T')
plt.ylabel('N')
plt.show()
```
