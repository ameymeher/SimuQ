```python
from qiskit import QuantumCircuit, schedule

qc = QuantumCircuit(2)

qc.rx(0.5,0)
qc.rzx(0.5,0,1)
qc.measure_all()

qc.draw('mpl')
```

```python
from qiskit_ibm_runtime import QiskitRuntimeService, Sampler

service = QiskitRuntimeService(channel="ibm_quantum")
backend = service.backend("ibmq_mumbai")
```

```python
from qiskit_transpiler_old import get_pm

pm_old = get_pm(backend)
transpiled_qc_old = pm_old.run(qc)
transpiled_qc_old.draw('mpl')
```

```python
transpiled_qc_old.calibrations
```

```python
sched_old = schedule(transpiled_qc_old, backend)
sched_old.draw()
```

```python
from qiskit_transpiler import get_pm

pm_new = get_pm(backend)
transpiled_qc_new = pm_new.run(qc)
transpiled_qc_new.draw('mpl')
```

```python
transpiled_qc_new.calibrations
```

```python
sched_new = schedule(transpiled_qc_new, backend)
sched_new.draw()
```
